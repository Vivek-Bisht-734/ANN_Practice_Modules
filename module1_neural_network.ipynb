## Module 1: Neural Network?

```python
# Simple Neural Network
import torch
import torch.nn as nn

class SimpleNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(3, 4)
        self.layer2 = nn.Linear(4, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
    def forward(self, x):
        x = self.sigmoid(self.layer2(self.relu(self.layer1(x))))
        return x

model = SimpleNetwork()
x = torch.tensor([[5.0, 7.0, 3.0]])
output = model(x)
print(output)
```

```python
# Mini Exercise: Building a neuron from scratch using only Python
import math

def sigmoid(z):
    return 1 / 1 + math.exp(-z)

def artificial_neuron(inputs, weights, bias, activation = 'sigmoid'):
    assert len(inputs) == len(weights), 'Number of imputs must match number of weights'

    weighted_sum = 0
    for x, w in zip(inputs, weights):
        weighted_sum += x * w
    z = weighted_sum + bias
    print(f'Weighted sum (before bias): {weighted_sum:.4f}')
    print(f'Weighted sum (after bias): {z:.4f}')

    if activation == 'sigmoid':
        output = sigmoid(z)
    elif activation == 'step':
        output = 1 if z >= 0 else 0
    elif activation == 'relu':
        output = max(0, z)
    else:
        raise ValueError(f'Unknown Activation: {activation}')
    print(f'After activation {activation}: output = {output:.4f}')
    
    return output

# Test 1: Exam Pass Prediction
inputs = [2.0, 7.0, 3.0]
weights = [0.9, 0.1, 0.6]
bias = 3.0
output = artificial_neuron(inputs, weights, bias, activation = 'sigmoid')
print(f'\nPrediction (pass probability): {output:.4f}')
print(f'Interpretation: {"PASS" if output >= 0.5 else "FAIL"}')

# Test 2: Perceptron with step function (XOR attempt)
print('\n' + '==' * 25)

xor_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
expected = [0, 1, 1, 0]
weights = [1.0, 1.0]
bias = -0.5

for (x1, x2), exp in zip(xor_inputs, expected):
    z = x1 * weights[0] + x2 * weights[1] + bias
    got = 1 if z >= 0 else 0
    correct = 'Right' if got == exp else 'Wrong'
    print(f'({x1}, {x2})   {exp:>10}  {got:>10}  {correct:>10}')

print(f'\nThe perceptron gails on XOR - it cannot draw a straight line to separate these.')
print('A Single Perceptron Cannot Solve XOR Problem.')
```
