```python
# Dot Product
x = torch.tensor([5.0, 7.0, 3.0])
w = torch.tensor([0.8, 0.4, 0.5])

# Method1: manual dot product
z = (x[0] * w[0] + x[1] * w[1] + x[2] * w[2])
print(f'Manual Method: {z:.4f}')

# Mehtod2: torch.dot (for 1D vectors only)
z = torch.dot(x, w)
print(f'torch.dot Method: {z:.4f}')

# Method3: @ operator (matrix multiplication, works for all dims)
z = x @ w
print(f'@ operator: {z:.4f}')
```

```python
# Autograd : How it works
w = torch.tensor(2.0, requires_grad = True)
b = torch.tensor(1.0, requires_grad = True)
x = torch.tensor(3.0)

z = w * x + b
L = z ** 2
print(f'z = {z.item():.1f}')
print(f'L = {L.item():.1f}')

L.backward()
print(f'dL/dw = {w.grad:.2f}')
print(f'dL/db = {b.grad:.2f}')
```

```python
# Full Gradient Descent Step : Trying for one gradient descent update
x = torch.tensor(2.0)
y = torch.tensor(10.0)
w = torch.tensor(1.0, requires_grad = True)
lr = 0.1

for step in range(10):
    pred = w * x
    loss = (pred - y) ** 2
    loss.backward()

    with torch.no_grad():
        w -= lr * w.grad
    w.grad.zero_()

    print(f'Epoch = {step + 1}, Weight = {w.item():.4f}, Pred = {pred.item():.4f}, Loss : {loss.item():.4f}, Zero grad update = {w.grad if w.grad is not None else "zeroed"}')
print(f'\nFinal w = {w.item():.4f} (target: y/x = {y.item()/x.item():.1f})')
```

```python
# Mini Exercise: Building a 2-layer network using tensor operations (no nn.Module)
# Architecture: Input(4) → Hidden(8) → Output(1)
# Activation: ReLU on hidden, Sigmoid on output
# Loss: MSE
# Then doing ONE gradient descent step.

import torch

torch.manual_seed(42)
x = torch.randn(10, 16)
y = torch.randint(0, 2, (10, 1)).float()

w1 = torch.randn(8, 16) * 0.1
b1 = torch.zeros(8)
w1.requires_grad_(True)
b1.requires_grad_(True)

w2 = torch.randn(1, 8) * 0.1
b2 = torch.zeros(1)
w2.requires_grad_(True)
b2.requires_grad_(True)

lr = 0.01

z1 = x @ w1.T + b1
a1 = torch.clamp(z1, min = 0)

z2 = a1 @ w2.T + b2
a2 = torch.sigmoid(z2)

loss = ((a2 - y) ** 2).mean()  # mse
print(f'Loss Before Update: {loss.item():.4f}')

loss.backward()
print(f'w1.grad Shape: {w1.grad.shape}')

with torch.no_grad():
    w1 -= lr * w1.grad
    b1 -= lr * b1.grad
    w2 -= lr * w2.grad
    b2 -= lr * b2.grad

w1.grad.zero_()
b1.grad.zero_()
w2.grad.zero_()
b2.grad.zero_()

z1 = x @ w1.T + b1
a1 = torch.clamp(z1, min = 0)
z2 = a1 @ w2.T + b2
a2 = torch.sigmoid(z2)
loss2 = ((a2 - y) ** 2).mean()
print(f'Loss After Update: {loss2.item():.4f}')
print(f'Loss Decreased: {loss2.item() < loss.item()}')

print('\nShape Tracing')
print(f'Input X: {x.shape}')
print(f'Z1: {z1.shape}')
print(f'A1: {a1.shape}')
print(f'Z2: {z2.shape}')
print(f'A2: {a2.shape}')
```
